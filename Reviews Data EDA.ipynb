```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf, col
from pyspark.sql.types import StringType

def remove_urls(text):
    import re
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub('', text)

remove_urls_udf = udf(remove_urls, StringType())
df = df.withColumn("review_text wo_urls", remove_urls_udf(col("review_text")))

#df.show(3, vertical=True, truncate=False)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf, col
from pyspark.sql.types import StringType

def remove_symbols(text):
    symbols_to_remove = ['.', ',', '!', '@', '-', "'", '"']
    for symbol in symbols_to_remove:
        text = text.replace(symbol, '')
    return text

remove_symbols_udf = udf(remove_symbols, StringType())

df = df.withColumn("review_text_wo_symbols", remove_symbols_udf(col("review_text")))

#df.show(3, vertical=True, truncate=False)
```

```python
punctuation = re.compile(r'[-.?!,:;()|0-9]')
post_punctuation = []

def remove_punctuations(gold_word_tokenize):
    for words in gold_word_tokenize:
        word = punctuation.sub("", words)
        if len(word)>0:
            post_punctuation.append(word)
    return post_punctuation

def stopwords_removal(post_punctuation):
    
    from nltk.corpus import stopwords
    from nltk.tokenize import word_tokenize
    
    stop_words = stopwords.words('english')
    post_stop_words = []
    word_tokenize = word_tokenize(post_punctuation)
    for words in word_tokenize:
        words = words.lower()
        if words not in stop_words:
            post_stop_words.append(words)           
    return (" ".join(post_stop_words)).strip()

df = df.withColumn("review_text_cleaned", remove_punctuations(df.select('review_text')))
```
